```python
# Ajout de colonnes Region et Income group en faisant une jointure
dataMain = dataMain.merge(country[['Country Code', 'Region', 'Income Group']], how='inner', on='Country Code')
```

```python
# Mise en évidence de la valeur la plus récente
# Création d'une colonne "Most Recent" pour inscrire la valeur la plus récente de l'indicateur
dataMain.insert(loc=10, column='Most Recent', value=None, allow_duplicates=False)
```

```python
# Remplissage de la colonne "Most Recent"
# On remplace la valeur de 'Most Recent' par '2015' puis on utilise la propriété de Nan qui ne peut pas être égal 
#à lui même pour détecter si 'Most Recent' est une valeur manquante, dans ce cas on essaie avec la valeur de
# l'année précédente et ainsi de suite.
for ind in dataMain.index :
    dataMain.at[ind, 'Most Recent'] = dataMain.at[ind, '2015']
    if dataMain.at[ind, 'Most Recent']!=dataMain.at[ind, 'Most Recent']:
        dataMain.at[ind, 'Most Recent'] = dataMain.at[ind, '2014']
        if dataMain.at[ind, 'Most Recent']!=dataMain.at[ind, 'Most Recent']:
            dataMain.at[ind, 'Most Recent'] = dataMain.at[ind, '2013']
            if dataMain.at[ind, 'Most Recent']!=dataMain.at[ind, 'Most Recent']:
                dataMain.at[ind, 'Most Recent'] = dataMain.at[ind, '2012']
                if dataMain.at[ind, 'Most Recent']!=dataMain.at[ind, 'Most Recent']:
                    dataMain.at[ind, 'Most Recent'] = dataMain.at[ind, '2011']
                    if dataMain.at[ind, 'Most Recent']!=dataMain.at[ind, 'Most Recent']:
                        dataMain.at[ind, 'Most Recent'] = dataMain.at[ind, '2010']
```

```python
# Calcul de la population cible à partir des 3 premiers indicateurs cible et insertion dans le dataframe
# Création d'un dictionnaire qui sera insérée dans le dataframe
for countryCode in dataMain['Country Code'].unique():
    newRow = dict()
#Recherche de l'index du pays actuel dans country
    countryIndex = country[country['Country Code'].str.contains(countryCode)].index[0]
    newRow['Country Name']=country.at[countryIndex, 'Table Name']
    newRow['Country Code']=countryCode
    newRow['Indicator Name']='Target Population Total'
    newRow['Indicator Code']='TGT.POP.TOT'
    newRow['2010']=None
    newRow['2011']=None
    newRow['2012']=None
    newRow['2013']=None
    newRow['2014']=None
    newRow['2015']=None
# Valeur la plus récente
# Recherche des indicateurs du pays actuel dans data
    countryData = dataMain[dataMain['Country Code'].str.contains(countryCode)]
# Recherche des index des valeurs démographiques
    indicatorIndex = countryData[countryData['Indicator Code'].str.contains('SP.POP.TOTL')].index[0]
    popTot = countryData.at[indicatorIndex, 'Most Recent']
    indicatorIndex = countryData[countryData['Indicator Code'].str.contains('SP.POP.0014.TO.ZS')].index[0]
    pop0014 = countryData.at[indicatorIndex, 'Most Recent']
    indicatorIndex = countryData[countryData['Indicator Code'].str.contains('BAR.SEC.ICMP.15UP.ZS')].index[0]
    popSec = countryData.at[indicatorIndex, 'Most Recent']
# Calcul de l'indicateur
    targetPop = popTot * (1-pop0014/100) * popSec/100
    newRow['Most Recent']=targetPop 
    newRow['Region']=country.at[countryIndex, 'Region']
    newRow['Income Group']=country.at[countryIndex, 'Income Group']
# Insertion de la nouvelle ligne
    dataMain = dataMain.append(newRow, ignore_index=True)
```

Notre dataframe de travail est presque prêt, il a été restreint aux pays étudiés et un nouvel indicateur a été calculé à partir des indicateurs démographiques. Attention, du fait du calcul de cet indicateur à partir des valeurs les plus récentes de chaque indicateurs, il est tout à fait possible que ce calcul soit effectué avec des indicateurs venant d'années différentes.

# Score d'attractivité

Pour la suite de l'étude, nous allons ramener les indicateurs étudiés à 3 : l'indicateur de population cible, l'indicateur de revenus et l'indicateur d'accès à internet. Nous nous permettons cette opération car l'indicateur de population cible est issu d'un calcul mathématique fiable. On ne pourrait pas, notamment, le multiplier avec l'indicateur d'accès à internet car il y a probablement une corrélation entre le niveau d'étude et l'accès à internet, les populations aisées des pays à fortes inégalités ont probablement un accès significativement meilleur à l'éducation et à internet en raison de leur richesse. De fait multiplier ces deux indicateurs de façon brute serait biaisé et donnerait une population cible plus faible qu'en réalité, d'autant plus dans les pays ou au moins l'un de ces deux indicateurs est faible.

Ce type de biais ne se retrouve pas dans le calcul de l'indicateur de population cible tel qu'il est à ce stade.

L'indicateur de population totale est conservé pour pondérer les moyennes par région et groupe de revenu. L'indicateur de croisssance de population est conservé pour un usage ultérieur.

### Normalisation des données

Un indicateur est en $international, un représente une population en nombre de personnes et le dernier est un pourçentage de population. Les multiplier ensemble pour obtenir un indicateur tout en un représentant "le pouvoir d'achat annuel total de la population de plus de 15ans ayant accès à internet et ayant à minima un niveau secondaire incomplet" pourrait donner un indicateur unique permettant de comparer de façon simple tous les pays entre eux, il aurait en plus une signification concrète. Cependant, il y a de nombreux biais dans ce calcul comme évoqué plus haut car l'accès à l'éducation, à internet et le revenu sont fortement corrélés.

Nous ne pouvons pas non plus aggréger ces indicateurs en l'état pour en faire un score car ils sont d'unités différentes, un nombre de personne, un montant en $international et un pourçentage. Nous allons donc les normaliser à l'aide d'une transformation de Box-Cox puis leur appliquer un coefficient avant de les aggréger pour en faire un score.

```python
# Création d'un data frame simplifié
dataMainEpured = dataMain[['Country Code','Indicator Code','Most Recent']]
dataMainPivoted = dataMainEpured.pivot(index=['Country Code'], columns='Indicator Code')['Most Recent']
dataMainSimple = dataMainPivoted[['TGT.POP.TOT','NY.GNP.PCAP.PP.CD','IT.NET.USER.P2','SP.POP.TOTL','SP.POP.GROW']]

# Création d'un indicateur de population cible 2030 pour tenir compte de l'évolution de la population
# En approche simple on applique le taux de croissance le plus récent 15 fois (2015-2030) sur la population cible
dataMainEvo = dataMainSimple.copy()
dataMainEvo['TGT.POP.TOT.2030']= [(row[1][-5]*(1+row[1][-1]/100)**15) for row in dataMainSimple.iterrows()]
dataMainSimple = dataMainEvo

# Transformation de Cox-Box et intégration dans un DF
dataScaled = sklearn.preprocessing.PowerTransformer(method = 'box-cox').fit_transform(dataMainSimple[['TGT.POP.TOT','NY.GNP.PCAP.PP.CD','IT.NET.USER.P2','TGT.POP.TOT.2030']])
dataScaled = pd.DataFrame(dataScaled, columns=['TGT.POP.TOT.sc','NY.GNP.PCAP.PP.CD.sc','IT.NET.USER.P2.sc','TGT.POP.TOT.2030.sc'])
```

```python
# Rassemblement des données
dataMainSimple = dataMainSimple.merge(country[['Country Code','Table Name', 'Region', 'Income Group']], how='inner', on='Country Code')
dataMainSimple = dataMainSimple.merge(dataScaled, how='inner', left_index=True, right_index=True)
dataFinal = dataMainSimple.copy()
```

### Calcul du score

Le score est basé sur les 3 indicateurs après transformation. Nous pourrions fixer à ce stade des coefficients mais comment les déterminer ?

Revenus : paramètre très important car il permet d'évaluer les prix praticables
Taille de la population : paramètre très important car il permet d'évaluer la taille du marché
Accès internet : un accès internet est indispensable pour l'utilisation du service mais cet indicateur permet aussi de se donner une idée sur l'intégration d'internet dans la consommation. En effet, un pays à faible nombre d'usager est peut-être peu habitué à utiliser internet et les services qu'on y trouve en général et l'apprentissage par internet en particulier.

Chaque indicateur étant important dans l'étude, nous avons choisi de ne pas les pondérer.

```python
# Pondération des indicateurs pour le calcul du score
pondNet = 1
pondPop = 1
pondGNP = 1
```

```python
dataMainSimple.head()
```

```python
# Calcul du score et du score 2030
dataFinal = dataMainSimple.copy()
dataFinal['Rating']=[((row[1][-2]*pondNet + row[1][-3]*pondGNP + row[1][-4]*pondPop)/(pondNet+pondGNP+pondPop)) for row in dataFinal.iterrows()]
dataFinal['Rating2030']=[((row[1][-3]*pondNet + row[1][-4]*pondGNP + row[1][-2]*pondPop)/(pondNet+pondGNP+pondPop)) for row in dataFinal.iterrows()]
# Top 10
dataFinal.sort_values(by='Rating',ascending=False).head(10)
```
